# Finding critical point $p_c$

```python
import os
import json
import numpy as np
import seaborn as sns
import matplotlib.pyplot  as plt
import scipy.stats as stats
import glob
```

```python
sns.set()
```

```python
root = '../data/critical/'
fig_root = "../figures/"
# os.listdir(root)
```

```python
signature = ["SitePercolation_ps_v10", "SitePercolationRSBD_L1", "SitePercolationRSBD_L2"]
files = glob.glob(root + signature[2] + "*")
```

```python
files
```

```python
data = dict()
for file in files:
    with open(file) as f:
        line = f.readline()
        head = json.loads(line[1:])
        L = int(head['length'])
    pcs = np.loadtxt(file, usecols=(0,))
#     pc = np.mean(pcs)
    if L in data.keys():
        data[L] = np.append(data[L], pcs)
    else:
        data[L] = pcs
    pass
```

```python
avg = []
for L in data.keys():
    avg.append([L, np.mean(data[L])])
    pass
```

```python
avg = np.array(avg)
```

```python
def plot_first(data):
    # fig = plt.figure(dpi=200)

    plt.plot(data[:,0], data[:,1], 'o')
    # plt.title(r"$p_c$ vs $L$")
    plt.xlabel(r"$L$")
    plt.ylabel(r"$p_c$")
    filename = fig_root + "finding-pc.eps"
    # plt.savefig(filename)
```

```python
plot_first(avg)
```


### It is evident that increasing system size gets you closer to the critical point

It flows the relation

$(p(L)-p_c(\infty)) L ^{1/\nu} = const $

$(p(L)-p_c) \sim L ^{-1/\nu}$

$(p(L)-p_c(\infty)) \sim (L^{-1})^{1/\nu}$

In the limit $L\rightarrow \infty$, $p(L) = p(\infty)$

If $1/\nu$ is known then finding $p_c$ is very easy.

```python
one_by_nu = 0.75
# one_by_nu = 0.736
one_by_nu = 0.721
# one_by_nu = 1 # for testing
```

```python
def plot_linear(data):
    x = data[:,0]**(-one_by_nu)
    plt.plot(x, data[:,1], 'o')
    regress = stats.linregress(x, data[:,1])
    x_min = np.min(x) - 0.001
    x_max = np.max(x) + 0.001
    x = np.linspace(x_min, x_max, 10)

    y = x*regress[0] + regress[1]
    plt.plot(x, y)

    plt.xlabel(r"$L^{-1/\nu}$")
    plt.ylabel(r"$p_c$")
    print(regress)
    return regress
```

```python
regress = plot_linear(avg)
```

```python
print("intercept is the pc")
print(regress[1])
```
